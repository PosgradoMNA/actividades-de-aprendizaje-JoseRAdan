#<font color='black'> **Parte 1: Fundamentos de bases de datos**
1.	Fundamentos de bases de datos y para ciencia de datos.  
Las bases de datos continúan siendo una de las principales herramientas para el manejo de datos pues ofrecen muchas ventajas sobre otros formatos como archivos de texto, hojas de cálculo, imágenes, etc., y es por ello que han estado vigentes de hace varias décadas en casi cualquier organización que maneje información.
Comparadas con los formatos mencionados se pueden enumerar las siguientes ventajas:
-	Facilidad y flexibilidad de extracción de datos para análisis.
-	Facilidad de asociación de datos.
-	Confiabilidad y eficiencia en el almacenaje de datos.
-	Rapidez de procesamiento.
Para ello, las bases de datos guardan los datos en especie de matrices como las hojas de cálculo donde claramente se pueden apreciar los valores de múltiples campos (columnas) y que se acumulan en renglones por lo que es posible conocer los diferentes valores que tomado una columna y asociarlo con otras columnas incluso de otras tablas.
El tema de optimización y eficiencia de almacenaje, las bases de datos permiten normalizar la información, o lo que es lo mismo, evitar duplicidad de datos que tienen el mismo significado.
Por todo lo anterior, la ciencia de datos, que trata de extraer conocimiento desde los datos, aprovecha muchas de las ventajas que ofrecen las bases de datos para su proceso de extracción y análisis de los mismos.

2.	Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.
Los almacenes de datos son bases de datos que concentran datos de múltiples fuentes, ya sea de otras bases de datos, archivos y procesos de tal manera que dichos datos pueden se explotados por analistas desde dicho almacén usando herramientas adecuadas para ello.
Las data warehouses ofrecen las ventajas de las bases de datos y generalmente almacenan cantidad de información muy grande y por ello es que son una de las fuentes principales para la ciencia de datos.
El reto que enfrentan los almacenes de datos está en la obtención de datos, que al ser de diversas fuentes, implica que los datos también son de diferentes formatos, Nemo siempre of compatibles o fáciles de asociar con el de las otras fuentes.
Uno de las formas que se ha arraigado para asegurar una mayor homogeneidad de los datos es la llamada ETL (Extract, Transform, Load) que establece que una vez que se extraen lo datos de las fuentes, estos se transforman a un formato más conveniente y luego se cargan al datawarehouse.
Mientras este proceso ha dado resultados satisfactorios durante años ahora se habla de un proceso ELT (Extract, Load, Transform) donde se utiliza la tecnología de nube de vanguardia que tiene la capacidad de almacenar datos en su forma original y de procesarlos (transformarlos) según convenga.
La conclusión es que ya sea que se transformen antes o después de cargarlos, es innegable que en algún momento deberán transformarse y almacenarse en algún formato que sea el más conveniente para su mejor análisis y que la tecnología actual expande las posibilidades de hacerlo de diferentes formas.

##<font color='blue'>Sustitución media

*<font color='blue'>La sustitución media es cuando el valor medio de una variable se usa en lugar del valor de datos faltantes para esa misma variable. Esto ayuda a utilizar los datos recopilados en un conjunto de datos incompleto. El trasfondo teórico de la sustitución de la media es que la media es una estimación razonable para una observación seleccionada al azar de una distribución normal. Sin embargo, si los valores que faltan no son estrictamente aleatorios, especialmente en el caso de que la distribución de los datos esté muy sesgada, el método de sustitución de la media puede no ser apropiado y conducirá a un sesgo inconsistente. En este caso, la sustitución de la mediana será más adecuada. La sustitución de la mediana es cuando se usa el valor de la mediana de una variable en lugar del valor de datos faltantes de la misma variable.*
##<font color='blue'>Me parece muy conveniente para rellenar los datos faltantes de la X5, o sea la edad pues es una variable cuantitativa.

```python
#wm = ndf.Weight.mean()
wm = ndf.X5.mean()
wm
```

```python
ndf['X5'].fillna(value = wm, 
                    inplace = True)
ndf
```

##<font color='blue'>De los datos X2 a X4, estos son discretos, es decir, no se puede tomar un calculo matemático tipo real, pero si pudemos usar un Hot Code Imputation, el cual se define como:

#<font color='blue'> **Hot deck imputation**

<font color='blue'>A randomly chosen value from an individual in the sample who has similar values on other variables.
In other words, find all the sample subjects who are similar on other variables, then randomly choose one of their values on the missing variable.
One advantage is you are constrained to only possible values. In other words, if Age in your study is restricted to being between 5 and 10, you will always get a value between 5 and 10 this way.
Another is the random component, which adds in some variability. This is important for accurate standard errors.</font>

```python
ndf[['X2','X3','X4']] = ndf[['X2','X3','X4']].fillna(method='ffill', inplace=False)
ndf
```

##<font color='blue'>Después de aplicarlo vemos que ahora las columnas X2 a X5 ya no contienen datos perdidos.

```python
ndf.isnull().sum()
```
