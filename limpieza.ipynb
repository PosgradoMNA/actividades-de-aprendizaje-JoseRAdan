```python
ndf.dropna(how='all', inplace = True)
ndf
```

## <font color='yellow'>El método anterior muestra que no hay filas donde falten todos los valores.

Alternativamente: usamos Threshold. 

Mantenga solo las filas con al menos 2 valores que  **NO SEAN** `nan`


```python
ndf = df.copy()
ndf
```

```python
ndf.dropna(thresh=20, inplace = True) # In a row, it needs at least 4 nan values is needed, to maintain in df
ndf # in case of column  add   axis=1 
```

## <font color='yellow'>La instrucción anterior muestra que casi todas las observaciones tienen al menos 4 atributos con valores, pero creo que 4 es una cantidad poco relevante pues son 23 atributos posibles.

## <font color='yellow'>Si incrementamos el parámetro treshold a 20 entonces solamente 12 filas son eliminadas con lo que todavía se mantienen una gran cantidad de datos útiles.
