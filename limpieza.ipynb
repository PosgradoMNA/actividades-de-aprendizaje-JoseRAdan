## <font color='yellow'>Sustitución media

La sustitución media es cuando el valor medio de una variable se usa en lugar del valor de datos faltantes para esa misma variable. Esto ayuda a utilizar los datos recopilados en un conjunto de datos incompleto. El trasfondo teórico de la sustitución de la media es que la media es una estimación razonable para una observación seleccionada al azar de una distribución normal. Sin embargo, si los valores que faltan no son estrictamente aleatorios, especialmente en el caso de que la distribución de los datos esté muy sesgada, el método de sustitución de la media puede no ser apropiado y conducirá a un sesgo inconsistente. En este caso, la sustitución de la mediana será más adecuada. La sustitución de la mediana es cuando se usa el valor de la mediana de una variable en lugar del valor de datos faltantes de la misma variable.
## <font color='yellow'>Me parece muy conveniente para rellenar los datos faltantes de la X5, o sea la edad pues es una variable cuantitativa.

```python
#wm = ndf.Weight.mean()
wm = ndf.X5.mean()
wm
```

```python
ndf['X5'].fillna(value = wm, 
                    inplace = True)
ndf
```

## <font color='yellow'>De los datos X2 a X4, estos son discretos, es decir, no se puede tomar un calculo matemático tipo real, pero si pudemos usar un Hot Code Imputation, el cual se define como:

# **Hot deck imputation**

A randomly chosen value from an individual in the sample who has similar values on other variables.

In other words, find all the sample subjects who are similar on other variables, then randomly choose one of their values on the missing variable.

One advantage is you are constrained to only possible values. In other words, if Age in your study is restricted to being between 5 and 10, you will always get a value between 5 and 10 this way.

Another is the random component, which adds in some variability. This is important for accurate standard errors.

```python
ndf[['X2','X3','X4']] = ndf[['X2','X3','X4']].fillna(method='ffill', inplace=False)
ndf
```

## <font color='yellow'>Después de aplicarlo vemos que ahora las columnas X2 a X5 ya no contienen datos perdidos.

```python
ndf.isnull().sum()
```

## <font color='yellow'>Para el resto de los datos, al tener valores sobre el comportamiento del cliente sobre como va a pagando a través de tiempo, considero que son bastante relevantes pero por eso mismo, creo que imputar valores es una cuestión complicada, considero que al ser muy pocos los valores faltantes, lo mejor sería eliminar las filas en la que hay un algun dato nulo en dichas columnas.

```python
ndf.dropna(inplace = True)
ndf.isnull().sum()
```

```python
ndf
```
